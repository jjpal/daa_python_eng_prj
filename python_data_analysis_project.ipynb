### Investigate Anomaly

We are asked to investigate an incident that happened between May 31, 2017 at midnight & less than June 2, 2017.

First thing to do is filter the columns & rows to get the data between these two dates

Return the earliest and latest date of the data set

### Filter Dataframe

```
# filter dataframe, where the date is larger than May 31, 2017 at midnight & less than June 2, 2017. 

df_june = df[(df['date'] > "2017-05-31 23:59:59") & (df['date'] < "2017-06-02")].reset_index(drop=True)
df_june
```

### Important Columns

The % Iron Concentrate is the most important variable. But our engineer colleague tells us that the % Silica Concentrate, Ore Pulp pH, & Flotation Column 05 Level and date are all really important as well. 

```
important_cols = ['date', '% Iron Concentrate', '% Silica Concentrate',
                  'Ore Pulp pH', 'Flotation Column 05 Level']

df_june_important = df_june[important_cols]
df_june_important
```
